```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
```

```python
def iconv(image, kernel):
    return convolve(image, kernel, mode='reflect')
```

```python
def kgauss(sigma, half_width):
    size = 2 * half_width + 1
    x = np.linspace(-half_width, half_width, size)
    y = np.linspace(-half_width, half_width, size)
    x, y = np.meshgrid(x, y)
    gauss_2d = np.exp(-(x**2 + y**2) / (2 * sigma**2))
    return gauss_2d / gauss_2d.sum()
```

```python
image_path = 'flowers4.png' 
image = plt.imread(image_path)
if image.ndim == 3:  
    image = np.mean(image, axis=2)
image = image.astype(np.float64) / 255.0 
```

1) Leia  e  mostre  a  imagem  flowers4.png  em  tons  de  cinza  (grey)  e  formato 
double. 

```python
plt.figure()
plt.title("Imagem Original (tons de cinza)")
plt.imshow(image, cmap='gray')
plt.colorbar()
plt.show()
```

2) Gerar um kernel médio de tamanho 15x15 com volume unitário (ou seja, todos 
os elementos somam 1) e armazene na variável kU. Plote o Kernel 

```python
kernel_uniform = np.ones((15, 15), dtype=np.float64)
kernel_uniform /= kernel_uniform.sum()
plt.figure()
plt.title("Kernel Médio (15x15)")
plt.imshow(kernel_uniform, cmap='gray')
plt.colorbar()
plt.show()
```

3) Aplicar o kernel kU à imagem usando a função iconv (). Armazene a imagem 
resultante na variável imU.

```python
image_u = iconv(image, kernel_uniform)
plt.figure()
plt.title("Imagem com Kernel Médio (imU)")
plt.imshow(image_u, cmap='gray')
plt.colorbar()
plt.show()
```

4) Use a função kgauss () para gerar um filtro Gaussiano com sigma = 5 e com 
meia largura de 8; armazene na variável kG. Plote o Kernel

```python
kernel_gauss = kgauss(sigma=5, half_width=8)

plt.figure()
plt.title("Kernel Gaussiano (sigma=5, meia largura=8)")
plt.imshow(kernel_gauss, cmap='gray')
plt.colorbar()
plt.show()
```

5) Aplicar  o  kernel  kG  à  imagem.    Armazene  a  imagem  resultante  na  variável 
imG 

```python
image_g = iconv(image, kernel_gauss)

plt.figure()
plt.title("Imagem com Kernel Gaussiano (imG)")
plt.imshow(image_g, cmap='gray')
plt.colorbar()
plt.show()
```

6) Exiba  as  imagens  imU  e  imG.  Verifique  os  efeitos,  notadamente  a 
consequência  da  operação  não  isotrópica  na  imU,  onde  linhas  verticais  e 
horizontais são criadas. 
 

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Imagem com Kernel Médio (imU)")
plt.imshow(image_u, cmap='gray')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("Imagem com Kernel Gaussiano (imG)")
plt.imshow(image_g, cmap='gray')
plt.colorbar()
```
