```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "flowers9.png"  
image = cv2.imread(image_path)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
R, G, B = image_rgb[:, :, 0], image_rgb[:, :, 1], image_rgb[:, :, 2]
```

a. Fa√ßa os histogramas referentes aos planos R, G e B da imagem; 

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Imagem Original (RGB)")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(R.ravel(), bins=256, color='red', alpha=0.7)
plt.title("Histograma do Canal R")
plt.xlabel("Intensidade")
plt.ylabel("Frequ√™ncia")

plt.subplot(2, 2, 3)
plt.hist(G.ravel(), bins=256, color='green', alpha=0.7)
plt.title("Histograma do Canal G")
plt.xlabel("Intensidade")
plt.ylabel("Frequ√™ncia")

plt.subplot(2, 2, 4)
plt.hist(B.ravel(), bins=256, color='blue', alpha=0.7)
plt.title("Histograma do Canal B")
plt.xlabel("Intensidade")
plt.ylabel("Frequ√™ncia")

plt.tight_layout()
plt.show()
```

b. Plote a imagem RGB e as imagens de cada plano separadamente; 
 
 
 
 

```python
plt.figure(figsize=(10, 6))

plt.subplot(1, 3, 1)
plt.imshow(R, cmap='Reds')
plt.title("Plano R")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(G, cmap='Greens')
plt.title("Plano G")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(B, cmap='Blues')
plt.title("Plano B")
plt.axis('off')

plt.tight_layout()
plt.show()
```
