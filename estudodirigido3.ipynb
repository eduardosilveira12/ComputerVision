```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Carregar a imagem e converter para escala de cinza
image = cv2.imread('penguins.png', cv2.IMREAD_GRAYSCALE)

# 2. Definir os kernels de Sobel
Kv = np.array([[-1, -2, -1],
               [ 0,  0,  0],
               [ 1,  2,  1]])  # Kernel vertical
Ku = Kv.T  # Kernel horizontal (transposta de Kv)

# 3. Aplicar convoluo com os kernels
Iu = cv2.filter2D(image, -1, Ku)  # Gradiente horizontal
Iv = cv2.filter2D(image, -1, Kv)  # Gradiente vertical

# 4. Calcular a magnitude do gradiente
I = np.sqrt(Iu**2 + Iv**2)
I = (I / np.max(I) * 255).astype(np.uint8)  # Normalizar para faixa [0, 255]

# 5. Plotar as imagens
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Gradiente Horizontal (Iu)')
plt.imshow(Iu, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Gradiente Vertical (Iv)')
plt.imshow(Iv, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Imagem das Bordas (I)')
plt.imshow(I, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
