```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def ajustar_brilho(imagem, valor):
    return cv2.convertScaleAbs(imagem, alpha=1, beta=valor)

def ajustar_contraste(imagem, fator):
    return cv2.convertScaleAbs(imagem, alpha=fator, beta=0)

def negativo(imagem):
    return cv2.bitwise_not(imagem)

def posterizar(imagem, niveis):
    fator = 256 // niveis
    return (imagem // fator) * fator

imagem_uint8 = cv2.imread('lena.pgm', cv2.IMREAD_GRAYSCALE)
imagem_double = imagem_uint8.astype(np.float64) / 255.0

brilho = ajustar_brilho(imagem_uint8, 50)
contraste = ajustar_contraste(imagem_uint8, 1.5)
negativa = negativo(imagem_uint8)
posterizada = posterizar(imagem_uint8, 4)

titulos = ['Original', 'Brilho', 'Contraste', 'Negativo', 'Posterizao']
imagens = [imagem_uint8, brilho, contraste, negativa, posterizada]

plt.figure(figsize=(12, 6))
for i in range(len(imagens)):
    plt.subplot(1, len(imagens), i+1)
    plt.imshow(imagens[i], cmap='gray')
    plt.title(titulos[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
