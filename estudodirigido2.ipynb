```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Carregar a imagem em tons de cinza
imagem = cv2.imread('castle.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Calcular o histograma
histograma = cv2.calcHist([imagem], [0], None, [256], [0, 256])

# 3. Identificar o nível de cinza mais frequente
nivel_mais_frequente = np.argmax(histograma)

# 4. Inspecionar visualmente o histograma
plt.figure(figsize=(8, 5))
plt.plot(histograma, color='black')
plt.title('Histograma de Intensidades')
plt.xlabel('Níveis de Cinza')
plt.ylabel('Frequência')
plt.axvline(x=nivel_mais_frequente, color='red', linestyle='--', label=f'Nível mais frequente: {nivel_mais_frequente}')
plt.legend()
plt.show()

# 5. Escolher manualmente um threshold
threshold_manual =  180

# 6. Criar imagem binária usando o threshold manual
imagem_binaria = (imagem > threshold_manual).astype(np.uint8) * 255

# 7. Mostrar a imagem binária
plt.figure(figsize=(6, 6))
plt.imshow(imagem_binaria, cmap='gray')
plt.title(f'Imagem Binária (Threshold Manual = {threshold_manual})')
plt.axis('off')
plt.show()
```
